```python
#import libraries
from tropycal import tracks
```

```python
#Read the data, a 'TrackDataset' object, hurdat is default, but use it anyway, don't use current year
basin = tracks.TrackDataset(basin='north_atlantic',source='hurdat',include_btk=False)
```

```python
## Get info for Sally
storm = basin.get_storm(('sally',2020))
print(storm)
```

```python
# Dictionary of data
print(storm.to_dict())
```

```python
#xarray dataset
print(storm.to_xarray())
```

### The following plots model data of temperature and salinity at sea surface and bottom.

```python
#pandas dataframe
print(storm.to_dataframe())
```

```python
#Plot...this takes a while
storm.plot()
```
